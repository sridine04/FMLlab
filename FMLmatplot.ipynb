**HISTOGRAM**

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("students grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7)
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("students grade")
plt.hist(marks,grade_intervals,histtype="step",rwidth=0.7)
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("students grade")
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="r")
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

```python
marks=[90,95,20,35,70,75,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("students grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7,facecolor="g")
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no of students")
plt.show()
```

**PIE CHART**

```python
import matplotlib.pyplot as plt
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values)
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance)
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=90)
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.3,0,0,0))
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.2,0.2,0.2,0.2))
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.2,0.2,0.2,0.2),shadow=True)
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.2,0.2,0.2,0.2),colors=["pink","yellow","purple","skyblue"])
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.2,0.2,0.2,0.2),colors=["pink","yellow","purple","skyblue"])
plt.legend(title="students_performance")
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.2,0.2,0.2,0.2),colors=["pink","yellow","purple","skyblue"],autopct="%2.1f%%")
plt.show()
```

```python
students_performance =('Excellent','Good','Average','Poor')
students_values =[35,30,20,15]
plt.figure(figsize=(9,10))
plt.pie(students_values,labels =students_performance,startangle=0,explode=(0.2,0.2,0.2,0.2),colors=["pink","yellow","lightgreen","skyblue"],autopct="%2.1f%%")
plt.legend(title="students_performance")
plt.show()
```
